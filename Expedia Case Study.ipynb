```python
import pandas as pd
import numpy as np
import os
from datatile.summary.df import DataFrameSummary
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
print(os.path.abspath(os.curdir))
path = os.path.abspath(os.curdir)
```

```python
df = pd.read_excel(path + '/Datasets/Dataset- (Data Science- Analytics).xlsx')
df.head()
```

```python
## This is our dfSummary table

## dfSummary uses the DataTile package to publish statistical measures for our pre-processed dfTable

dfSummary = DataFrameSummary(df)
dfSummary.summary()
```

```python
## We observe there are many values missing in the super-region column

NaN_df = df[df.isna().any(axis = 1)] 
NaN_df
```

```python
## We take the summary of the rows with NaN values in Super Region and observe our country name has 1 unique value.

NaN_dfs = DataFrameSummary(NaN_df)
NaN_dfs.summary()
```

```python
## Let's see the distribution of Super-Region NaN values by Property Country

NaN_Vals = NaN_df.groupby('Property Country')['Net Orders', 'Net Gross Booking Value USD'].sum()
NaN_Vals
```

```python
## Let's calculate the booking value per order of the records with NaN in Super Region

NaN_Vals['Booking value per order'] = NaN_Vals['Net Gross Booking Value USD']/NaN_Vals['Net Orders']
NaN_Vals.sort_values(by = 'Net Gross Booking Value USD', ascending=False)
```

```python
## After analyzing this data on excel, we realize that NA was interpeted by pandas as NaN.
## Let's go ahead and change the Super Region from NA to NaN 

df['Super Region'] = df['Super Region'].fillna('NAM')
df.head()
```

```python
## We were having trouble with our groupby function, let's go ahead and clean some more rows that can't be 
## addressed by our calculations

indexes_to_drop = df[(df['Net Gross Booking Value USD'] <= 0) | (df['Net Orders'] <= 0)].index
df = df.drop(index = indexes_to_drop, axis = 0)
df.head()
```

```python
## Let's normalize our Net Gross Booking Value so we can easily plot the colors in Tableau

book_value_scale = preprocessing.MinMaxScaler(feature_range= (0, 1))
book_vals = [[item] for item in df['Net Gross Booking Value USD'].values]
df['Net Gross Booking Value USD Scaled'] = book_value_scale.fit_transform(book_vals)
df.head()
```

```python
## Let's add a Booking Value per order

df['Booking value per order'] = df['Net Gross Booking Value USD']/df['Net Orders']
df.sort_values(by = 'Net Gross Booking Value USD', ascending=False)
df.head()
```

```python
## We observe the week with the most net bookings independent of negative net orders means to the dataset, 
## was 2017-W45.

## This will be our dfWeek Table
## dfWeek contains sum of Net Gross Booking Value USD over each week.

dfWeek = df.groupby('Week')['Net Gross Booking Value USD'].sum()
dfWeek.sort_values(ascending = False)
```

```python
## In order to analyze our dataset more clearly by region and mobile indicator name, let's group by these two 
## columns, while summing over our Net Orders and Net Gross Booking Value USD, and take the mean of our Net Gross

## This will be our dfRegion Table
## dfRegion conatins sum of Net Orders, Sum of Net Gross Booking Value USD, and the mean of Booking value per order 
## across Mobile Indicator Names (Desktop/Mobile), and Super Regions

aggregators = {'Net Orders':'sum','Net Gross Booking Value USD': 'sum', 'Booking value per order' : 'mean'}
dfRegion = df.groupby(['Super Region','Mobile Indicator Name'], dropna=False).agg(aggregators)
dfRegion.head()
```

```python
## Here we observe that the highest value customer segment for Expedia are those bookings done 90+ days in advance. 
## This is a good indication that our experiment in PCA could be functional.

## This will be our dfBookingGroup Table
## dfBookingGroup contains sum of Net Orders, Sum of Net Gross Booking Value USD, and the mean of Booking value per order 
## across Booking Window Groups

dfBookingGroup = df.groupby(['Booking Window Group']).agg(aggregators)
dfBookingGroup.sort_values(by = 'Booking value per order', ascending = False)
dfBookingGroup.head()
```

```python
## Let's set up our dataframe to be able to calculate growth rates for each country week-over-week

## This will be our df_country_growth_rate
## df_country_growth will group countries, sum of Net Orders, sum of Net Gross Booking Value USD, and mean of Booking
## value per order across weeks. This will be helpful when calculating our week-over-week growth rates

df_country_growth = df.groupby(['Week', 'Property Country']).agg(aggregators)
df_country_growth.head()
```

```python
## Here we groupby each country to calculate pct_change across our metrics

df_country_growth['Net Order Percent Change'] = df_country_growth.groupby(['Property Country'])['Net Orders'].pct_change()
df_country_growth['Net Gross Booking Value USD Percent Change'] = df_country_growth.groupby(['Property Country'])['Net Gross Booking Value USD'].pct_change()
df_country_growth['Booking value per order Percent Change'] = df_country_growth.groupby(['Property Country'])['Booking value per order'].pct_change()
df_country_growth
```

```python
## Let's check our calculations

df_country_growth.loc[slice(None), 'Pakistan', :]
```

```python

## Let's print out the period and country with the most highest percent growth of Net Orders
highest_net_order_change = df_country_growth.loc[df_country_growth['Net Order Percent Change'].idxmax()]
print(highest_net_order_change.name)
print(highest_net_order_change)

## Let's print out the period and country with the most highest percent growth of Net Gross Booking Value USD
highest_gross_booking_val_change = df_country_growth.loc[df_country_growth['Net Gross Booking Value USD Percent Change'].idxmax()]
print(highest_gross_booking_val_change.name)
print(highest_gross_booking_val_change)

## Let's print out the period and country with the most highest percent growth of Booking Value per order
highest_gross_booking_value_per_order = df_country_growth.loc[df_country_growth['Booking value per order Percent Change'].idxmax()]
print(highest_gross_booking_value_per_order.name)
print(highest_gross_booking_value_per_order)
```

```python
## Let's parition our data into the first unique group across Property Country

df_country_growth_first = df_country_growth.reset_index().set_index('Week').set_index('Property Country')
df_country_growth_first = df_country_growth_first.groupby('Property Country').first()
df_country_growth_first
```

```python
## Let's parition our data into the first unique group across Property Country

df_country_growth_last = df_country_growth.reset_index().set_index('Week').set_index('Property Country')
df_country_growth_last = df_country_growth_last.groupby('Property Country').last()
df_country_growth_last
```

```python
## Let's left join the first table with the right table, which allows us to make calulations for Total Growth 
## column-wise 

## df_country_growth_first_last contains the Total Growth rates for each country's Net Orders, Net Gross Booking
## Value USD, and Booking value per order. These calculations are done independent of the consistnecy of a 
## countries last booking period

df_country_growth_first_last = pd.merge(df_country_growth_first, df_country_growth_last, on = 'Property Country')
df_country_growth_first_last
```

```python
## Let's reassign relevant columns, and drop the unnecessary columns from df_country_growth_first_last 

df_country_growth_first_last['Net Order Percent Change'] = df_country_growth_first_last[['Net Orders_x', 'Net Orders_y']].pct_change(axis='columns')['Net Orders_y']
df_country_growth_first_last['Net Gross Booking Value USD Total Percent Change'] =  df_country_growth_first_last[['Net Gross Booking Value USD_x', 'Net Gross Booking Value USD_y']].pct_change(axis='columns')['Net Gross Booking Value USD_y']
df_country_growth_first_last['Booking Value per order Percent Change'] = df_country_growth_first_last[['Booking value per order_x', 'Booking value per order_y']].pct_change(axis='columns')['Booking value per order_y']
df_country_growth_first_last = df_country_growth_first_last.drop(df_country_growth_first_last.iloc[:, 0:12], axis = 1)
```

```python
## This is our df_mobile_vs_desktop_country Table

## df_mobile_vs_desktop_country group's records by Week, Property Country, and Platform Type, aggregating the sum of
## Net Orders, sum of Net Gross Booking Value USD, and mean of Booking value per order.

df_mobile_vs_desktop_country = df.groupby(['Week', 'Property Country', 'Platform Type Name']).agg(aggregators)
df_mobile_vs_desktop_country
```

```python
df_mobile_vs_desktop_country['Net Order Percent Change by Platform'] = df_mobile_vs_desktop_country.groupby(['Property Country', 'Platform Type Name'])['Net Orders'].pct_change()
df_mobile_vs_desktop_country['Net Gross Booking Value USD Percent Change by Platform'] = df_mobile_vs_desktop_country.groupby(['Property Country', 'Platform Type Name'])['Net Gross Booking Value USD'].pct_change()
df_mobile_vs_desktop_country['Booking value per order Percent Change by Platform'] = df_mobile_vs_desktop_country.groupby(['Property Country', 'Platform Type Name'])['Booking value per order'].pct_change()
df_mobile_vs_desktop_country
```
